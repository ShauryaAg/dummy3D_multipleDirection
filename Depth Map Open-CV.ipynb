```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

plt.rcParams['figure.figsize'] = (16, 9)
plt.axis('off')

imgL = cv2.imread('./New Ones/new_left (4).jpg', cv2.IMREAD_GRAYSCALE)
imgR = cv2.imread('./New Ones/new_right (4).jpg', cv2.IMREAD_GRAYSCALE)


(image_height, image_width) = imgL.shape[:2]

resized_width = int(image_width/3)
resized_height = int(image_height/3)
dim = (resized_width, resized_height)

imgL = cv2.resize(imgL, dim, interpolation=cv2.INTER_AREA)
imgR = cv2.resize(imgR, dim, interpolation=cv2.INTER_AREA)

# print(imgR.shape)
# print(imgL.shape)

# img_array_left = np.asarray(imgL, dtype="int32")
# img_array_right = np.asarray(imgR, dtype="int32")
stereo = cv2.StereoBM_create(numDisparities=96, blockSize=5)
disparity = stereo.compute(imgL, imgR)
# cv2.imshow('image', disparity)
plt.imshow(disparity, cmap='inferno')
# plt.clim(0, 500)
plt.colorbar()
plt.show()
```
